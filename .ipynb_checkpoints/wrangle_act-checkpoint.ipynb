```python
tweet_ids = list(twitter_df.tweet_id)

tweet_data = {}
for tweet in tweet_ids:
    try:
        tweet_status = api.get_status(tweet, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
        tweet_data[str(tweet)] = tweet_status._json
        
    except:
        print("Error for: " + str(tweet))
        continue

#THIS TOOK 4EVA
```

### Code

```python
def replace_names(row):
    # replace dog's named a
    if row['name'] == 'a':    
        #get text from row
        text = row['text']
        
        # word list from text
        word_list = text.split()

        # defaults to no name
        name = 'None'

        # get name from 'named {}'
        if 'named' in word_list:
            # next word after 'named'...+1 will give us next word which is name
            name_pos = word_list.index('named') + 1
            name = word_list[name_pos]

        # get name from 'name is {}'
        elif 'name is' in text:
            #  next word after 'name is'...we want to bypass (is). +2 will omit is and give us name
            name_pos = word_list.index('name') + 2
            name = word_list[name_pos]

        #only replace the 'name' column (with no period)
        row['name'] = name.replace('.','')
    
    # Return row whether updated or name
    return row
    

twitter_clean = twitter_clean.apply(replace_names, axis=1)
```

While we are at it...we can also condense the predictions from the images dataframe into one column as well

### Code

```python
images_clean.sample(4)
```

```python

breeds = []
confidence = []

def breeds_confidence(row):
    if row['p1_dog']:
        breeds.append(row['p1'])
        confidence.append(row['p1_conf'])
    elif row['p2_dog']:
        breeds.append(row['p2'])
        confidence.append(row['p2_conf'])
    elif row['p3_dog']:
        breeds.append(row['p3'])
        confidence.append(row['p3_conf'])
    else:
        breeds.append('Unidentifiable')
        confidence.append(0)

images_clean.apply(breeds_confidence, axis = 1)

images_clean['breeds'] = breeds
images_clean['confidence'] = confidence
```

```python
# now we can remove the redundant rows 
images_clean.drop(['p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog',], axis=1, inplace=True)
```

```python
def val_counts(column_name):
    
    frequency = dftwitter[column_name].value_counts()
    
    return frequency
```

```python
def groupby_sum(column_name1, column_name2):
    
    grp = dftwitter.groupby(column_name1)[column_name2].sum()
    
    return grp
```

```python
def groupby(column_name1, column_name2):
    
    grp = dftwitter.groupby(column_name1)[column_name2]
    
    return grp
```

```python
def groupby_avg(column_name1, column_name2):
    
    grp = round(dftwitter.groupby(column_name1)[column_name2].mean(),2)
    
    return grp
```
